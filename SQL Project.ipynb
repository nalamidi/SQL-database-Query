# Join(Combine) "Restaurant Table" with "Food Table"

# Change restaurant 'name' to 'restaurant'

# Join "Restaurant Table" with "Food Table" and "Categories Table"

```python
sql = """
SELECT foods.id, foods.name, restaurants.name as restaurant, categories.name as category
FROM foodS
INNER JOIN restaurants ON foods.restaurant_id = restaurants.id
INNER JOIN categories ON foods.category_id = categories.id
"""
pd.read_sql_query(sql, engine).head()
```

```python
sql = """
SELECT f.id, f.name, r.name as restaurant, c.name as category
FROM foods f
INNER JOIN restaurants r ON f.restaurant_id = r.id
INNER JOIN categories c ON f.category_id = c.id
WHERE r.id = 1
"""
pd.read_sql_query(sql, engine).head()
```

#  Groupby Average calories per restaurant_id

```python

sql = """
SELECT f.restaurant_id, AVG(f.calories)
FROM foods f
INNER JOIN restaurants r on r.id = f.restaurant_id
GROUP BY (f.restaurant_id)
"""
pd.read_sql_query(sql, engine).head()
```

# Groupby Average calories per restaurant

```python

sql = """
SELECT f.restaurant_id, r.name, AVG(f.calories)
FROM foods f
INNER JOIN restaurants r on r.id = f.restaurant_id
GROUP BY (f.restaurant_id, r.name)
"""
pd.read_sql_query(sql, engine).head()
```

# Groupby "Restaurant_id" and "Name"
# Count calories per restaurant

```python
sql = """
SELECT f.restaurant_id, r.name, COUNT(f.calories)
FROM foods f
INNER JOIN restaurants r on r.id = f.restaurant_id
GROUP BY (f.restaurant_id, r.name)
"""
pd.read_sql_query(sql, engine).head()
```

# Groupby Minimum calories per restaurant

```python

sql = """
SELECT f.restaurant_id, r.name, MIN(f.calories)
FROM foods f
INNER JOIN restaurants r on r.id = f.restaurant_id
GROUP BY (f.restaurant_id, r.name)
"""
pd.read_sql_query(sql, engine).head()
```

# To use groupby
# Sum calories per restaurant

```python
sql = """
SELECT f.restaurant_id, r.name, SUM(f.calories)
FROM foods f
INNER JOIN restaurants r on r.id = f.restaurant_id
GROUP BY (f.restaurant_id, r.name)
"""
pd.read_sql_query(sql, engine).head()
```

# AVG calories per restaurant

```python
sql = """
SELECT c.name, AVG(f.calories)
FROM categories c
INNER JOIN foods f on f.category_id = c.id
GROUP BY (c.name)
ORDER BY c.name
"""
pd.read_sql_query(sql, engine).head()
```

# Join multiple tables
# Select only a condition where categories name is "Burger"
# Restaurant name is "McDonald's" or "Burger King"

```python
sql = """
SELECT f.name, c.name as category, r.name as restaurant, f.calories
FROM foods f
INNER JOIN categories c ON c.id = f.category_id 
INNER JOIN restaurants r ON f.restaurant_id = r.id
WHERE c.name = 'Burgers'
AND (r.name = 'McDonald''s' OR r.name = 'Burger King')
ORDER BY f.name
"""
pd.read_sql_query(sql, engine).head()
```

# Categories name is "Kid's Meals"

```python
sql = """
SELECT f.name, f.calories, c.name as category
FROM foods f
INNER JOIN categories c ON c.id = f.category_id
INNER JOIN restaurants r ON r.id = f.restaurant_id
WHERE c.name = 'Kid''s Meals'
ORDER BY f.calories
"""
pd.read_sql_query(sql, engine).head()
```

```python
sql = """
SELECT r.name, MAX(f.calories)
FROM foods f
INNER JOIN restaurants r on f.restaurant_id = r.id
INNER JOIN categories c on c.id = f.category_id
WHERE c.name = 'Kid''s Meals'
GROUP BY (r.name)
ORDER BY MAX(f.calories) DESC
"""
pd.read_sql_query(sql, engine).head()
```
